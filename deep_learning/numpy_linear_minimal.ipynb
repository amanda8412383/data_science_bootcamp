```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

### lineare Regression, minimal Beispiel

```python
# input generieren
observation = 1000
variables = 2
input = np.random.uniform(-10, 10, (observation, variables)) 
input.shape
```

```python
#gewicht generieren
weight = np.random.uniform(-10, 10, (variables, 1))
weight.shape
```

```python
# target generieren
noise = np.random.uniform(-1, 1, (observation, 1))
#noise soll nicht zu groß sein, deshalb (1, -1)
bias = np.random.uniform(-10, 10, (1, 1))
target = np.dot(input, weight) + bias + noise
```

```python
#Datei plotten
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.plot(input[:, 0], input[:, 1], target.reshape(observation, ))
#lineare Beyiehung in einem 3D-Diagramm ist eine Ebene
```

```python
#Variablen initialisieren
init_range = 0.5
init_weight = np.random.uniform(-init_range, init_range, (2, 1))
init_bias = np.random.uniform(-init_range, init_range, 1)
learning_rate = 0.1
```

```python
#Model tranierem
for i in range(100):
    output = np.dot(input, init_weight) + init_bias
    delta = output - target
    #L2-norm loss Funktion
    loss = np.sum(delta ** 2)
    #loss per Stichproben und Variablem, daher die Lengeschwindigkeit ist unabhängig von der Anzahl der Stichproben
    loss_scaled = loss/observation/variables
    delta_scaled = delta/observation
    #Grdientenabstiegslogik
    init_weight = init_weight - learning_rate * np.dot(input.T, delta_scaled)
    init_bias = init_bias - learning_rate * np.sum(delta_scaled)
    print(loss_scaled)
```
